```python
import numpy as np
```

```python
a = np.arange(15).reshape(3,5)
print(a)
print(f'a.shape: {a.shape}') # ndarray 의 배열의 모양
print(f'a.ndim: {a.ndim}') # ndarray 의 차원
print(f'a.dtype.name: {a.dtype.name}') # ndarray 의 데이터 타입
print(f'a.itemsize: {a.itemsize}') # ndarray 의 바이트 크기
print(f'a.size: {a.size}') # ndarray 의 총개수
print(f'type(a): {type(a)}') # ndarray 의 타입
```

```python
a= np.array([1,2,3,4]) # np.array(list)
print(a)
print(type(a))
print(a.dtype)


b = np.array([1.5,2.5,5.1])
print(b)
print(type(b))
print(b.dtype)
```

```python
c = np.array([[1,2,3],[4,5,6]])
print(c)
```

```python
d = np.array([[1,2],[3,4]] , dtype = complex)
print(d)
```
