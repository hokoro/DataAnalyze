```python
import numpy as np

a = np.arange(6).reshape((2,3))
b = np.zeros_like(a)
```

```python
print(a)
print(b)
```

```python
c = np.arange(6).reshape((2,3))
d = np.ones_like(c)
```

```python
print(c)
print(d)
```

```python
e = ([1,2,3],[4,5,6])
e_like = np.empty_like(e)

print(e)
print(e_like)
```

```python
f = ([1.,2.,3.] , [4.,5.,6.])
f_like = np.empty_like(f)
```

```python
print(f)
print(f_like)
```

```python
g = np.full((2,3),np.inf)
print(g)

h = np.full((2,3),5)
print(h)
```

```python
i = np.arange(5 , dtype = int)
i_full_like = np.full_like(i,5) # dtype 에 맞는 형태로 full_like 수행

```
